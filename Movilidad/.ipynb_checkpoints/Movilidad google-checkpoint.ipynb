```python
def tabla_semana_zona(BD_gupeada,
                       indicador,
                       nombre_grupo,
                       nombre_fecha):
    dum1 = BD_gupeada[indicador].unstack()
    dum1.keys().name = nombre_fecha
    dum1.index.name = nombre_grupo
    dum1.columns = fecha_semanal
    return dum1
```

Defino una funciòn para el cálculo de los distintos indicadores para cada una de la regiones


```python
def tablamedia_semanaporregion(BD_gupeada,
                              nombre_region,
                              guardar=False,
                              fecha_semanal=None):
    dum1 = BD_gupeada.xs(nombre_region,level = 'sub_region_1')
    dum1.columns = categorias
    
    # en caso de que indique, se apendan los nombres de las semanas
    if fecha_semanal != None:
        dum1.index = fecha_semanal
    
    # en caso de que indique, se guarda archvo
    if guardar == True:

        # un dictado de las regiones y sus numeros
        num_region = mapa_regiones[['cir_sena','Region']].set_index('Region').to_dict()['cir_sena']
        
        print('Guardado:' + './Producto/figtab/tab_media_region' + str(num_region[str(nombre_region)]) + '.tab',)
        tabla = open('./Producto/figtab/tab_media_region' + str(num_region[str(nombre_region)]) + '.tab','w')
        print(dum1.to_latex(bold_rows= True,
                            float_format="%.1f",
                            column_format = 'rcccccc',
                            col_space=0.01),
              file=tabla)
        tabla.close()
    return dum1

# tablamedia_semanaporregion(chile_grouped_media,
#                       'Arica y Parinacota',
#                      fecha_semanal = fecha_semanal,
#                      guardar=True).head()
```

```python
def tablaDS_semanaporregion(BD_gupeada,
                              nombre_region,
                              guardar=False,
                              fecha_semanal=None):
    dum1 = BD_gupeada.xs(nombre_region,level = 'sub_region_1')
    dum1.columns = categorias
    
    # en caso de que indique, se apendan los nombres de las semanas
    if fecha_semanal != None:
        dum1.index = fecha_semanal
    
    # en caso de que indique, se guarda archvo
    if guardar == True:

        # un dictado de las regiones y sus numeros
        num_region = mapa_regiones[['cir_sena','Region']].set_index('Region').to_dict()['cir_sena']
        
        print('Guardado:' + './Producto/figtab/tab_DS_region' + str(num_region[str(nombre_region)]) + '.tab',)
        tabla = open('./Producto/figtab/tab_DS_region' + str(num_region[str(nombre_region)]) + '.tab','w')
        print(dum1.to_latex(bold_rows= True,
                            float_format="%.1f",
                            column_format = 'rcccccc',
                            col_space=0.01),
              file=tabla)
        tabla.close()
    return dum1

# tablaDS_semanaporregion(chile_grouped_DS,
#                       'Arica y Parinacota',
#                      fecha_semanal = fecha_semanal,
#                      guardar=True)
```

```python
def graficomedia_regiones(BD_groupeada,
                          nombre_region,
                          guardar=False):

    print(nombre_region)
    region_actual = tablamedia_semanaporregion(BD_groupeada,
                                          nombre_region)

    region_actual.plot().legend(bbox_to_anchor=(1, 1),
                                 title= 'Categoría')

    plt.xlabel('Fecha')
    plt.ylabel('Media regional \n indicador de movilidad')
    plt.title(
    'Media de movilidad ciudadana en base a servicios Google \n \
    y localización basada en el teléfono móvil \n \
    Región: ' + nombre_region)
    plt.xticks(sems,
               fecha_semanal,
               rotation='vertical')
    plt.ylim([-80, 35])
    
    if guardar == True:
        #dictado region y numero
        num_region = mapa_regiones[['cir_sena','Region']].set_index('Region').to_dict()['cir_sena'] 
        
        print('Guardado:' + './Producto/figtab/Regionmedia_'+ str(num_region[str(nombre_region)]) +'.png')
        plt.savefig('./Producto/figtab/Regionmedia_'+ str(num_region[str(nombre_region)]) +'.png',
                    bbox_inches='tight',
                    dpi = 150)
```

```python
dum1 = chile_grouped_media.transit_stations.unstack()
dum1.index.name = 'Región'

geo = mapa_regiones[['Region', 'geometry']].set_index('Region')
data = dum1.iloc[:,-1]
data.name = 'col'

dum2 = pd.concat( [geo, data], axis=1)


# Función para centrar en el mapa una región
def limites_mapas(BD, zona=None):
    lista = []
    if zona != None:
        lista = [zona]
        limites = dum2.iloc[dum2.index.isin(lista)== True][['geometry']].bounds

    else:
        lista = ['Valparaíso']
        limites = dum2.iloc[dum2.index.isin(lista)== False][['geometry']].bounds
    maps_lim = pd.concat(
        [limites.min()[['minx', 'miny']],
         limites.max()[['maxx', 'maxy']]
        ])
    return maps_lim

limites_mapas(dum2, 'Aysén')
```
